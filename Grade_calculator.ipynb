```python
print("Enter Marks Obtained in 5 Subjects: ")
Maths = int(input("Enter Marks Obtained in Maths: "))
Stats = int(input("Enter Marks Obtained in Stats: "))
DataScience = int(input("Enter Marks Obtained in datascience: "))
English = int(input("Enter Marks Obtained in English: "))
Sanskrit = int(input("Enter Marks Obtained in Sanskrit: "))
total = Maths+Stats+DataScience+English+Sanskrit
avg = total / 5
if 0 <= avg <= 100:
  if avg >= 90:
    print("Your grade is o")
  elif avg >= 80:
    print("Your Grade is A")
  elif avg >= 70:
    print("Your Grade is A+")
  elif avg < 60:
    print("Your Grade is B")
  elif avg < 70:
    print("Your Grade is B+")
  elif avg < 60:
    print("Your Grade is B")
  elif avg < 50:
    print("Your Grade is C+")
  elif avg < 40:
    print("Your Grade is C")
  elif avg < 30:
    print("Your Grade is D")
  else:
    print("Your Grade is F")
  print("Grade: ",avg)
```
